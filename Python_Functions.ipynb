# Return Statement

```python
def list_total(lst):
    """
    The function prints sum of a given list.
    Input:List of integers.
    Output:Sum of given list of integers
    """
    total=0
    for i in lst:
        total+=i
    return total
```

```python
total=list_total([1,2,3])
print(total)
```

```python
lst=[11,22,33,44,55]
list_total(lst)
```

# Arbitary Function Arguments/VARargs:

```python
greet('Sai','Good evening','test')
```

```python
def greeting(*names):
    for name in names:
        print(f"Hello {name}")
```

```python
greeting()
```

```python
greeting('Sai')
```

```python
greeting('MSD','RD','VK')
```

# Recursive Functions

```python
#Factorial
def get_factorial(num):
    return 1 if num<=1 else num*get_factorial(num-1)
```

```python
fact=get_factorial(5)
```

```python
fact
```

```python
#Lambda and filter.
#print positive numbers.
lst=list(range(-10,10))
lst
```
