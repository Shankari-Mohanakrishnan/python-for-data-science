# List comprehension 

```python
#obtain square root of a list.
x=[1,2,3,4,5]
y=[]
for i in x:
    y.append(i*i)
print(y)
    
```

```python
x=[1,2,3,4,5]
y=[i*i for i in x]
print(y)
```

```python
#Print vowles in a sentence.
sen="Today is an amazing day!"
vowels="aeiou"
e=[]
for i in sen:
    if i in vowels:
        e.append(i)
print(e)
```

```python
sen="Today is an amazing day!"
vowels="aeiou"
res=[i for i in sen if i in vowels]
res
```

# Python Operations

```python
a=set({1111,2222,3333,4444,5555})
a
```

```python
b=set({4444,5555,7777,8888,9999})
b
```

```python
#Intesection - returns only common values in both the sets
a.intersection(b)
```

```python
#union - return unique values in both the sets.
a.union(b)
```

```python
#Difference - returns the elements in the first list that are not there in the second list.
a.difference(b)
```

```python
#symmentric_differene - removes common elements from both the list and returns the other values
a.symmetric_difference(b)
```

# Programs:

To check if a given string is a palindrome or not.

```python
user_input=input("Enter a string to check if the string is a palindrome or not: ")
temp=user_input
reversed_str=user_input[::-1]
if user_input==reversed_str:
    print(f"{user_input} is a palindrome")
else:
    print(f"{user_input} is not a palindrome")
    
```
