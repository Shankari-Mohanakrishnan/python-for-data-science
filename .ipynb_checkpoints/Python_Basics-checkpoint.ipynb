# Output Formatting

```python
p=555
q=333
print("p is {} and q is {}".format(p,q))
```

Postional and Keyword Arguments:

```python
#postional aruguments: We have to maintain the order of arguments in .format()
print("p is {0} and q is {1}".format(p,q))
```

```python
#Keyword arugment: The order specified in .format() need not be maintained.
print("The story is about {first} and {second}".format(second="Ram",first="Sai"))
```

```python
#combining positional and keyword arguments:
print("Today is the {0} of {month}".format("13th",month="october"))
```

```python
#// - floor division divide the number and round to closest number
num = int(input("Enter a number: "))
order = len(str(num))
#print(order)
temp=num
tot=0
while temp>0:
    digit=temp%10
    tot+=digit**order
    temp//=10
print(tot)
print(num)
if num == tot:
    print("Amstrong number")
else:
    print("Not an amstrong number")
    
    
```

```python
# Changed num variable to string, 
# and calculated the length (number of digits)
num = int(input("Enter a number: "))
order = len(str(num))

# initialize sum
sum = 0

# find the sum of the cube of each digit
temp = num
while temp > 0:
   digit = temp % 10
   sum += digit ** order
   temp //= 10

# display the result
if num == sum:
   print(num,"is an Armstrong number")
else:
   print(num,"is not an Armstrong number")
```
